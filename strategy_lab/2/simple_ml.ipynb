```python
import os
import pandas as pd
import numpy as np

# Change the working directory to strategy_lab
os.chdir("/Users/sugang/Desktop/projects/2024coin/strategy_lab")

# Now, you can read the CSV file without specifying the full path
df = pd.read_csv("1/full_data_with_predictions.csv", index_col=0)
df
```

```python
df["signal"] = 0
df["signal"] = np.where(df["predict_close"] > df["close"].shift(1), 1, 0)
df["strategy_returns2"] = df["signal"] * ((df["close"] * 0.998 - df["open"] * 1.002) / df["open"] * 1.002)
df["cumulative_returns2"] = (1 + df["strategy_returns2"]).cumprod()
df.to_csv("2/temp2.csv")
```
