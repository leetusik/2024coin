```python
import pandas as pd
import numpy as np

df = pd.read_excel("data/btc_search_data.xlsx")
df
```

```python
df["ma10"] = df["score"].rolling(window=10).mean()
```

```python
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the columns 'Date', 'score', and 'price2'

# Plotting the data
plt.figure(figsize=(14, 7))

# Plot 'score' over time
plt.plot(df['Date'], df['score'], label='Score', color='blue', linewidth=2)

# Plot 'price2' over time
plt.plot(df['Date'], df['price2'], label='Price2', color='green', linewidth=2)

plt.plot(df['Date'], df['ma10'], label='ma10', color='red', linewidth=2)

# Adding titles and labels
plt.title('Score and Price2 Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Values', fontsize=14)

# Adding a legend
plt.legend()

# Adding gridlines for better readability
plt.grid(True)

# Rotating the x-axis labels for better readability
plt.xticks(rotation=45)

# Displaying the plot
plt.tight_layout()
plt.show()
```
